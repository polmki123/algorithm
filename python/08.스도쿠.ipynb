```python
from collections import *
import copy

''''
0 3 5 4 6 9 2 7 8
7 8 2 1 0 5 6 0 9
0 6 0 2 7 8 1 3 5
3 2 1 0 4 6 8 9 7
8 0 4 9 1 3 5 0 6
5 9 6 8 2 0 4 1 3
9 1 7 6 5 2 0 8 0
6 0 3 7 0 1 9 5 2
2 5 8 3 9 4 7 6 0

1 3 5 4 6 9 2 7 8
7 8 2 1 3 5 6 4 9
4 6 9 2 7 8 1 3 5
3 2 1 5 4 6 8 9 7
8 7 4 9 1 3 5 2 6
5 9 6 8 2 7 4 1 3
9 1 7 6 5 2 3 8 4
6 4 3 7 8 1 9 5 2
2 5 8 3 9 4 7 6 1

[1, 3, 5, 4, 6, 9, 2, 7, 8]
[7, 8, 2, 1, 3, 5, 6, 4, 9]
[4, 6, 9, 2, 7, 8, 1, 3, 5]
[3, 2, 1, 5, 4, 6, 8, 9, 7]
[8, 7, 4, 9, 1, 3, 5, 2, 6]
[5, 9, 6, 8, 2, 7, 4, 1, 3]
[9, 1, 7, 6, 5, 2, 3, 8, 4]
[6, 4, 3, 7, 8, 1, 9, 5, 2]
[2, 5, 8, 3, 9, 4, 7, 6, 1]
'''
full_it = deque()

n = [ list( [0] for i in range(9) ) for i in range(9) ]
for i in range(9) :
    for j, data in enumerate( map( int, input().split() ) ) :
        n[i][j] = data
        if data == 0 :
            full_it.append( (i,j) )

def eight_check( n, cord ) :
    x, y = cord[0], cord[1]
    point = {
        'count' : 0,
        'sum' : 0
    } 
    p_x = x//3
    p_y = y//3
    for i in range(p_x * 3, p_x * 3 + 3) :
        for j in range(p_y * 3, p_y * 3 + 3) :
            if ( i != x or j != y ) :
                if n[ i ] [ j ] != 0 :
                    point['count']  += 1
                    point['sum']    += n[ i ] [ j ]
                else :
                    return False

    if point['count'] == 8 :
        rest = 45 - point['sum']
        n[x][y] = rest 
        return True

    return False

def col_check( n, cord ) :
    point = {
        'count' : 0,
        'sum' : 0
    } 
    for i in range(9) :
        if cord[1] != i :
            if n[cord[0]][i] != 0 :
                point['count']  += 1
                point['sum']    += n[ cord[0]] [ i ]
            else :
                return False

    if point['count'] == 8 :
        rest = 45 - point['sum']
        n[cord[0]][cord[1]] = rest 
        return True

    return False

def row_check( n, cord ) :
    point = {
        'count' : 0,
        'sum' : 0
    } 
    for i in range(9) :
        if cord[0] != i :
            if n[i][cord[1]] != 0 :
                point['count']  += 1
                point['sum']    += n[i][cord[1]]
            else :
                return False

    if point['count'] == 8 :
        rest = 45 - point['sum']
        n[cord[0]][cord[1]] = rest 
        return True

    return False
 
while full_it :
    cord = full_it.popleft()
    if eight_check(n, cord) == False and row_check(n, cord) == False and col_check(n,cord) == False :
        full_it.append(cord)
    # if eight_check(n, cord) == True :
    #     for i in range(9):
    #         print(n[i])
    #     break

for i in range(9):
    print(*n[i])
```
