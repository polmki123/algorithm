```python
def curry(func) :
    def curried (*args, **kwargs) :
        def f(*args2, **kwargs2) :
            nonlocal args, kwargs 
            args += args2
            kwargs.update(kwargs2)
            return curried(*args, **kwargs)

        if len(args) + len(kwargs) >= func.__code__.co_argcount :
            result = func(*args, **kwargs)
            args, kwargs = [], {}
            return result 
        else :
            return f
    return curried

@curry
def reduce(f, acc) :
    acc = iter(acc)
    initate = next(acc)
    for a in acc :
        initate = f(initate, a)
    return initate

@curry 
def map(f, iterable) :
    result = []
    for i in iterable :
        result.append(f(i) )
    return result 


def go(*args):
    def func(a, f) :
        return f(a)
    return reduce(func, args)

```
